**The goal of the project is to perform data wrangling on the dataset og Automobile cars**

data wrangling is a process of :


1.   cleaning the data by removing null values and duplicates
2.   processing the data by creating dummy values



```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import the dataset

automobile_data = pd.read_csv('/content/sample_data/Automobile_data.csv')
automobile_data.head(10)
```

Identifying and handling missing data in the dataset

```python
automobile_data.replace('?',np.nan, inplace = True)
automobile_data.head()
```

Evaluating missing data

```python
missing_data = automobile_data.isnull()
missing_data.head()
```

```python
automobile_data.isnull().sum()
```

```python
for column in missing_data.columns.values.tolist():
  print(column)
  print(missing_data[column].value_counts())
  print('')
```

Replace the missing data in each column with its mean

Calculate the mean for normalized-losses.

```python
avg_norm_loss = automobile_data['normalized-losses'].astype('float').mean(axis=0)
print('Average of normalized-losses:',avg_norm_loss)
```

```python
#replace nan values with the mean 
automobile_data['normalized-losses'].replace(np.nan, avg_norm_loss, inplace = True)
```

```python
#the mean value for bore 
avg_bore = automobile_data['bore'].astype('float').mean(axis=0)
print('Average of num-of-doors:',avg_bore)
```

```python
automobile_data['bore'].replace(np.nan, avg_bore,inplace= True)
```

```python
avg_stroke = automobile_data['stroke'].astype('float').mean(axis=0)
print('Average of stroke:',avg_stroke)
automobile_data['stroke'].replace(np.nan,avg_stroke,inplace= True)
```

```python
avg_horsepower = automobile_data['horsepower'].astype('float').mean(axis=0)
print('Average of horsepower:',avg_horsepower)
automobile_data['horsepower'].replace(np.nan, avg_horsepower,inplace=True)
```

```python
avg_peak_rpm = automobile_data['peak-rpm'].astype('float').mean(axis=0)
print('Average of peak-rpm:',avg_peak_rpm)
automobile_data['peak-rpm'].replace(np.nan,avg_peak_rpm,inplace=True)
```

```python
automobile_data['num-of-doors'].value_counts()
```

We can see that four doors are the most common type. We can also use the ".idxmax()" method to calculate the most common type automatically:



```python
automobile_data['num-of-doors'].value_counts().idxmax()
```

```python
##replace the missing 'num-of-doors' values by the most frequent 

automobile_data['num-of-cylinders'].replace(np.nan,'four',inplace=True)
```

```python
# simply drop whole row with NaN in "price" column
automobile_data.dropna(subset=['price'],axis=0, inplace = True)

# reset index, because we droped two rows
automobile_data.reset_index(drop = True,inplace=True)
```

```python
automobile_data.head(10)
```

```python
automobile_data.dtypes
```

```python
#convert the datatype

automobile_data[['bore','stroke']] = automobile_data[['bore','stroke']].astype('float')
automobile_data[['normalized-losses']] = automobile_data[['normalized-losses']].astype('int')
automobile_data[['price']] = automobile_data[['price']].astype('float')
automobile_data[['peak-rpm']] = automobile_data[['peak-rpm']].astype('float')
```

```python
automobile_data.dtypes
```

**Data Standardization**

Data is usually collected from different agencies in different formats. (Data standardization is also a term for a particular type of data normalization where we subtract the mean and divide by the standard deviation.)



```python
automobile_data.head()
```

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)

automobile_data['city-L/100km'] = 235/automobile_data['city-mpg']

automobile_data.head()
```

Data Normalization


```python
# replace (original value) by (original value)/(maximum value)

automobile_data['length'] = automobile_data['length']/ automobile_data['length'].max()
automobile_data['width'] = automobile_data['width']/ automobile_data['width'].max()
```

```python
automobile_data['height'] = automobile_data['height']/ automobile_data['height'].max()

automobile_data[['length','width','height']].head()
```

**Binning**


```python
automobile_data['horsepower'] = automobile_data['horsepower'].astype(int, copy=True)
```

```python
%matplotlib inline

plt.hist(automobile_data['horsepower'])

plt.xlabel('horsepower')
plt.ylabel('count')
plt.title('Horsepower bins')
```

```python
bins = np.linspace(min(automobile_data['horsepower']),max(automobile_data['horsepower']),4)
bins
```

```python
groups_name = ['Low','Medium','High']
```

```python
automobile_data['horsepower-binned'] = pd.cut(automobile_data['horsepower'], bins, labels= groups_name,include_lowest=True)
automobile_data[['horsepower','horsepower-binned']].head(20)
```

```python
automobile_data['horsepower-binned'].value_counts()
```

```python
plt.bar(groups_name, automobile_data['horsepower-binned'].value_counts())
plt.xlabel('horsepower')
plt.ylabel('count')
plt.title('horsepower bins')
```

**Indicator Variable (or Dummy Variable)**

```python
automobile_data.columns
```

```python
dummy_variable_1 = pd.get_dummies(automobile_data['fuel-type'])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'diesel':'fuel-type-diesel','gas':'fuel-type-gas'},inplace=True)
dummy_variable_1.head()
```

```python
automobile_data = pd.concat([automobile_data,dummy_variable_1],axis=1)

automobile_data.drop('fuel-type', axis = 1,inplace=True)

automobile_data.head()
```

```python
dummy_variable_2 = pd.get_dummies(automobile_data['aspiration'])

dummy_variable_2.rename(columns={'std':'aspiration-std','turbo':'aspiration-turbo'},inplace=True)
dummy_variable_2.head()
```

```python
automobile_data = pd.concat([automobile_data, dummy_variable_2], axis=1)

automobile_data.drop('aspiration',axis=1,inplace=True)
```
