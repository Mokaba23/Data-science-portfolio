**Exploratory Data Analysis with Pandas**

```python
import pandas as pd
import numpy as np
```

We going to be analyzing  the dataset  on churn rate of telecom operator clients.
#churn rate - is the rate at which customers stop doing business with an entity.

```python
#read the data

df = pd.read_csv('/content/telecom_churn.csv')
df.head()
```

each row(instance) represent a single customer records and the columns are characteristics.

```python
#lets see the size of the dataframe
df.shape
```

```python
#check the columns
df.columns
```

```python
#check each column datatype
df.dtypes
```

```python
#conver the  churn data type to an integer type
df['Churn'] = df['Churn'].astype('int64')
```

```python
#statistics on the columns
df.describe()
```

```python
#lets see the statistics of non numerical values
df.describe(include =['object','bool'])
```

```python
#check for null values
df.isnull().sum()
```

```python
#lets look at the distribution of churn
df['Churn'].value_counts()
```

2850 clients/users out of 3333 are loyal their Churn values 0.To calculate fractions pass normalize = True

```python
df['Churn'].value_counts(normalize=True)
```

```python
#arrange total day charge
df.sort_values('Total day charge', ascending= False).head()
```

```python
df.sort_values(['Churn','Total day charge'], ascending = [True,False]).head()
```

Indexing and retrieving data

```python
#calculate the column standard deviation
df['Churn'].std()
```

```python
#calculate the column the mean
df['Churn'].mean()
```

```python
#average of churned users
df[df['Churn'] == 1].mean()
```

```python
#how much on average does churned users spend time on the phone during the day?
df[df['Churn']==1]['Total day minutes'].mean()
```

What is the maximum length of international calls among loyal users (Churn == 0) who do not have an international plan?



```python
df[df['Churn']==0 & (df['International plan']== 'No')]['Total intl minutes'].max()
```

```python
# loc is used for names and iloc for numeral values
#we print from row 0 to 5 , and from State column to Area code column, using dataframe indexing.
df.loc[0:5, 'State':'Area code']
```

```python
df.iloc[0:5, 0:3]
```

```python
#reverse indexing
df[-1:]
```

```python
df.apply(np.max)
```

```python
#find states that starts with W
df[df['State'].apply(lambda state: state[0] == 'W')].head()
```

```python
d = {'No':False, 'Yes': True}
df['International plan'] = df['International plan'].map(d)
df.head()
```

```python
df = df.replace({'Voice mail plan':d})
df.head()
```

Grouping data

```python
columns = ['Total day minutes','Total eve minutes','Total night minutes']
df.groupby(['Churn'])[columns].agg([np.mean,np.std,np.min,np.max])
```

Summary tables

```python
pd.crosstab(df['Churn'], df['International plan'])
```
